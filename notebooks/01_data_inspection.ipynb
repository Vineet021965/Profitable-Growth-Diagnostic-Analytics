## Customers Dataset — Initial Observations

- The dataset contains 99,441 rows and 5 columns.
- Each row represents a customer record associated with an order.
- `customer_id` is unique for every row and can be used to connect customers with the orders dataset.
- `customer_unique_id` represents the actual customer and may appear multiple times when the same customer places multiple orders.
- The dataset contains no missing values.
- No exact duplicate rows were found.
- Customer location information is available through city, state, and ZIP code prefix.
- For customer-level metrics such as total unique customers, repeat purchase rate, and customer profitability, `customer_unique_id` should be used instead of `customer_id`.

## Orders Dataset — Initial Observations

- The dataset contains 99,441 order records and 8 columns.
- Each row represents one order, making `order_id` the primary key candidate for the Orders table.
- `customer_id` connects the Orders dataset to the Customers dataset.
- Every `customer_id` in Orders has a matching record in the Customers dataset, indicating valid referential integrity between the two tables.
- No exact duplicate rows were found.
- Most orders have a `delivered` status, while a smaller number are canceled, unavailable, shipped, invoiced, processing, created, or approved.
- Missing values exist mainly in approval and delivery timestamp columns.
- These missing values should not be removed automatically because some are associated with legitimate business states such as canceled, unavailable, or undelivered orders.
- The dataset covers purchases from September 2016 to October 2018.
- Before profitability analysis, timestamp columns will need to be converted to datetime data types.
- A business rule will be required to determine which order statuses should be included in revenue and profitability calculations.

## Order Items Dataset — Initial Observations

- The dataset contains 112,650 rows and 7 columns.
- Each row represents one item position within an order.
- `order_id` is not unique because one order can contain multiple item rows.
- `order_item_id` is a sequence number within each order and is not globally unique.
- The combination of `order_id` and `order_item_id` uniquely identifies each row, making it a composite key candidate.
- No exact duplicate rows or missing values were found.
- Every `order_id` in the Order Items dataset has a matching record in the Orders dataset, indicating valid referential integrity from Order Items to Orders.
- Some orders do not have corresponding item records. These orders should be investigated by order status before deciding how they should be handled.
- Product prices and freight values are non-negative, although the distributions contain high-value observations that should be investigated before treating them as outliers.
- Most orders contain one item row, while a smaller number contain multiple item rows.
- The Orders to Order Items relationship is one-to-many.
- `price` represents the item selling price, while `freight_value` represents the freight amount associated with that item row.

## Products Dataset — Initial Observations

- Each row represents one unique product in the product catalog.
- `product_id` uniquely identifies each product and is the primary key candidate.
- No exact duplicate rows were found.
- Missing values exist in product category and several product attribute columns.
- Missing product attributes should not be removed automatically because products with incomplete metadata may still have valid transactions.
- Every `product_id` referenced by Order Items exists in the Products dataset, indicating valid referential integrity from Order Items to Products.
- The Products to Order Items relationship is one-to-many because the same product can appear in multiple item rows across different orders.
- Product purchase frequency is highly uneven: some products are purchased many times while many products have relatively few purchases.
- Product category, weight, and dimensions may be useful later for product-mix, freight-cost, and profitability analysis.
- Missing categories will require a documented handling rule before category-level analysis.
